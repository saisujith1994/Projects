## Exploring the features of Delta lake

```python
%sql
describe history deltalake_demo
```

```python
%sql
SELECT * 
FROM deltalake_demo VERSION AS OF 3
```

```python
%sql
SELECT * FROM deltalake_demo@v2
```

```python
%sql
CREATE or REPLACE TABLE deltalake_demo2
USING DELTA
PARTITIONED BY (age)
AS
SELECT * FROM deltalake_demo
```
