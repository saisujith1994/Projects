Function to get sample from all tables inside database, format of database value:
catalog.database

```python
def SampleAllTables(n, database):
    table_list = []
    df = spark.sql(
        f"show tables in {database}"
    )
    for i in range(df.count()):
        table_list.append(df.collect()[i][1])
    for table in table_list:
        sample_df = spark.sql(
            f"select * from {database}.{table} limit {n}"
        )
        display(table)
        display(sample_df)
```

```python
SampleAllTables(3,'samples.accuweather')
```
